## 2nd Dimension: Databases

Database is a collection of data manage by DBMS. DBMS manage data by defining **schema** (it's definition for the data) and **data files** (contains data).

When data is manage under DBMS, it needs to be a Schema. A schema is define using a data model. The relations model is the most common model. The schema can be describe using Entitiy Relationship (ER) model. Schema has data constraints to make sure data integrity. Constraint ensure the data keep is meaningfull.

### What is that schema & data model?


So basically, schema define/describe the data.

A schema is a blueprint of the database which specifies what fields will be present and what would be their types.  For example an employee table will have an employee_ID column represented by a string of 10 digits and an employee_Name column with a string of 45 characters.

A data model is a language for describing data or information which generally consit of:

1. structure of the data
2. Operations on the class
3. Constraints on the data.


Data model is a high level design which decides what can be present in the schema. It provides a database user with a conceptual framework in which we specify the database requirements of the database user and the structure of the database to fulfill these requirements.

A data model can, for example, be a relational model where the data will be organised in tables whereas the schema for this model would be the set of attributes and their corresponding domains.
    
### Popular Data Model
    
1. The relational model for structure data, include object-relational extension.


2. The semi-structure data model, inlcuding XML and related standards.

### Difference Between Schema and Data Model

The difference is that a **data model describes requirements** for data definitions of descriptive information and relationships, while a database, either with a schema or schemaless, is the code that implements storage on a specific database platform.

So more simply the **data model is the definition**, the **database schema is the implementation**. The data model is the design for the data irrespective of storage technology; the database schema is the design for the data modified to fit in a specific database technology.

#### Meta Data

Meta data/data dictionary is data describing the database which include:

1. Schema definition
2. Index (index speed up query processing)
3. Data types
4. COnstraints on data
5. User informations and privileges

## Transactions and Concurrency Control

DBMS has many user that operate on the database concurrenly. Transcations in an atomic unit consist of one or more operations. All of the operations on the transaction must be completed or nothing happens at all.

The execution of transactions should be durable, even there is system failure. We wanted to ensure that all the transactions is executed. To achive this we need **loggin**

### ACID Properties of Transactions

1. Atomicity - All or nothing executions of transactions

2. Consistency - A transactions bring database from one state to another. EX bank balance can't be negative

3. Isolation - Each transactions must appear to execute as if no other transactions are occuring

4. Durability - Effect of a transaction must neber be lost once a transaction completed
