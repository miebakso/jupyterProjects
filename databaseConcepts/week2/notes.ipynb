## Data Models

2. Operation on the data

## Data Integrity Constraint - Key and Primary Key

A **primary key** and **key** are **data integrity constraint** to ensure meaningful data. A key can be an attribute or a set of attribute to distinguish one row/tuples from another. In the real world there can be the same movies with the same title, so using one attribute is not enough to distinguish 2 tuples.

if there are **several candidate keys**, one is specified as the **Primary Key**. 



Set of attributes can be used to distinguish any 2 tuples in the relation. The key movieID in Movie can be used to distinguised 2 movies. In that case it doesn't matter if 2 movies have the same title, they must have 2 different movieID. Although moiveID is artifical, not like title or other attributes. 

Sometimes using one atrifical attribute is better than using multiples attributes to be the key. Yes, we can use title+date+director as keys to distinguish one movie from another, but it's too much. 

**There are situation Where 1 key is not enough**

For instance the movie "Angles and Demon" have 2 genre which is action and drama. In this case because of the relation, we need to have 2 row which consist of row with the genre action and another with the genre drama. So to distinguish 2 rows with another, movieID is not enough. Genre itself is not enough because there can be another movie with the same genre. 

When you combine (movieID, Genre) it is enough to distinguish one tuples/row from another. It's called **composite key**, it is a candidate key that consists of two or more attributes.
