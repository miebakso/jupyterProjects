## Data Science Process

These are generals steps to data science:

1. Defining the research goals

2. Retrieving data

3. Data preparation

4. Data Exploration

5. Data modeling

6. Presentation and automation


The process of data science may note be linear, but itterative. Sometimes we skip some steps or repeat steps. 


### 1. Defining the research goals

We need to **understand** the **purpose** of the project/research. Ask question to make sure/clarify the research goals. Important questions:

- What is the project expected to do?
- Why is value being placed on the project?
- Hod should the analysis be carried out?

In this first step, we need to get an **agreement** on **deliverables** throught a **project charter**, which may include:

- Statement of research goals
- Broader project mission and context
- Required resources and data
- How analysis will be performed
- Proof that it's an achievable project
- Measure of sucess
- FOrmal deliverable (e.g.project / report)


### 2. Identify and Retrieve Data

Identify the data we need, how to retrieve data. Is that data internal or external. This step maybe challenging because of access to data (physical or digital barrier). Some data is sensitive and valuable so getting it may not be easy. Data maybe in raw form and required preprocessing.


Looking data at external data maybe very usefull in some projects.

### 3. Data Preparation

Data preprocessing/preparation/wrangling is a process where we are getting the data ready for the next step. it may include:

- **Data cleaning**:
    
    - Errors from data entry
    - Physically impossbile value
    - Missing value
    - Outlier
    - Typos, spaces, ...
    - Erorrs against codebook
  

- **Data Transformation**:
    
    - Aggregating data
    - Extrapolating data
    - Derived measures
    - Creating dummies
    - Reducing number of variables
    
- **Combining Data**:

    - Joining Data
    - Set Operatos
    - Creating Views
    
    
Why do we need to prepare data? data may contains errors.outlier/etc and these error may ruin the result. **Garbage In = Garbage Out**, no matter how good the model is, the result will be garbage if you put garbage.
